```python
ap.dropna(inplace=True)
```

```python
labelencode = LabelEncoder()
```

```python
textcol =['Name','Email Address','Username','Gender','Location','Subscription Plan','Payment Information','Renewal Status','Usage Frequency','Purchase History','Favorite Genres','Devices Used','Engagement Metrics']
```

```python
for col in textcol:
    ap[col] = labelencode.fit_transform(ap[col])
```

```python
ap.dtypes
```

```python
ap.drop(['Name', 'Email Address', 'Username', 'Date of Birth', 'Membership Start Date', 'Membership End Date'],axis=1,inplace=True)
```

```python
ap.head()
```

```python
X=ap[['Subscription Plan','Payment Information','Usage Frequency','Favorite Genres']]
```

```python
KMeans = KMeans(n_clusters=3, random_state=42)
```

```python

KMeans.fit(X)
```

```python
ap['cluster'] = KMeans.labels_
```

```python
ap.head()
```

```python
silhouette = silhouette_score(X,ap['cluster'])
```
